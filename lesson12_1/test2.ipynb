```python
from openpyxl import Workbook
from random import randint, choices

def getStudents(a: int) -> list[dict]:
    students = []
    with open('names.txt', mode='r', encoding='utf-8') as file:
        names = file.read()
    nameList = names.split('\n')
    names = choices(nameList, k=a)

    for i in range(a):
        stu = {
            '姓名': names[i],
            '國文': randint(0, 100),
            '英文': randint(0, 100),
            '數學': randint(0, 100),
            '地理': randint(0, 100),
            '歷史': randint(0, 100),
        }
        students.append(stu)

    return students

s_nums = int(input("請輸入學生的人數(1~50): "))
students = getStudents(s_nums)

wb = Workbook()
ws = wb.active
ws.title = "3年6班"
ws.append(['姓名', '國文', '英文', '數學', '地理', '歷史'])
for stu in students:
    ws.append(list(stu.values()))

ws1 = wb.create_sheet("3年5班")
ws2 = wb.create_sheet("3年4班")
wb.save("3年6班.xlsx")
```
